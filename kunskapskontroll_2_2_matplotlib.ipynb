### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

The two ways of plotting in Matplotlib is the explicit or object oriented way and the implicit or functional way where the explicit way is the recommended approach.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

A Figure is an object where a Matplotlib graph or diagram is drawn. The Figure, a window or Jupyter widgets for example, keep tabs of everything that the diagram is made of like lines, dots, bars, titles labels and so on can be specified in terms of x-y coordinates.
The graph or diagram it self is drawn on objects that is called Axes. A Figure can have one or many Axes. An Axes is an Artist attached to a Figure.
Axis is an object that set the scale and limits and generate ticks(the marks on the Axis).
An Artist is basically everything visible on the Figure. Like the Figure itself, Axes and Axis etc.


### When plotting in Matplotlib, what is the expected input data type?

Answer
The expected inpt when plotting in Matplotlib is numpy.array or numpy.ma.masked_array or objects that can be passed to numpy.asarray. Classes like pandas and numpy.matrix may not work as intended so the common convention is to convert these to numpy.array objects before plotting.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 10)
y = x**2

# Plotting the object oriented approach
fig, ax = plt.subplots(figsize=(5, 3), layout='constrained')
ax.plot(x, y)
ax.set_title('Object oriented approach')
ax.set_xlabel('x-label')
ax.set_ylabel('y-label')

# Plotting the pyplot approach
plt.figure(figsize=(5, 3), layout="constrained")
plt.plot(x, y)
plt.title("pyplot approach")
plt.xlabel("x-label")
plt.ylabel("y-label")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1, 2,  figsize=(10, 5))

axs[0].scatter(x=random_data_x, y=random_data_y)
axs[0].set_title("Random Dots")
axs[0].set_xlabel("X") 
axs[0].set_ylabel("Y")

axs[1].set_title("Frukter i lager")
axs[1].set_xlabel("Frukt") 
axs[1].set_ylabel("Antal")
axs[1].bar(x=names, height=values)
```
